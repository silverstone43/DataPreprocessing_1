```python
import pandas as pd
df = pd.read_csv("sample_1.csv", encoding = "utf-8-sig")
```

```python
df["입실일자"] = pd.to_datetime(df["입실일자"],format='%Y%m%d', errors='coerce')
df["입실시각"] = pd.to_datetime(df["입실시각"],format='%H%M', errors='coerce').dt.time
df["퇴실일자"] = pd.to_datetime(df["퇴실일자"],format='%Y%m%d', errors='coerce')
df["퇴실시각"] = pd.to_datetime(df["퇴실시각"],format='%H%M', errors='coerce').dt.time
df["접수일자"] = pd.to_datetime(df["접수일자"],format='%Y%m%d', errors='coerce')
df
```

```python
df = df.drop(columns=['검사코드'], errors='ignore')
df = df.drop(columns=['진료과코드'], errors='ignore')
df = df.drop(columns=['진료과코드NM'], errors='ignore')
df = df.drop(columns=['성별코드NM'], errors='ignore')
df
```

```python
df['접수시각'] = ""
df
```

```python
re = df["검사나이"]
result = pd.DataFrame(re)
result
```

```python
def categorize(result):
    if 81<=result:
        return "80세이상"
    elif 80 == result:
        return "80세"
    elif 70<=result<=79:
        return "70대"
    elif 60<=result<=69:
        return "60대"
    elif 50<=result<=59:
        return "50대"
    elif 40<=result<=49:
        return "40대"
    elif 30<=result<=39:
        return "30대" 
    elif 20<=result<=29:
        return "20대"
    elif 10<=result<=19:
        return "10대"
```

```python
result["검사나이"] = result["검사나이"].apply(categorize)
```

```python
df["검사나이"] = result["검사나이"]
df
```

```python
df["입원기간"] = df["퇴실일자"] - df["입실일자"]
df
```

```python
df = df.dropna(subset = ["기준치상한값"])
df = df.dropna(subset = ["기준치하한값"])
df
```

```python
df["환자ID"] = (df["연구번호"].astype(str).str[:3] + df["연구번호"].astype(str).str[5:])
df
```
