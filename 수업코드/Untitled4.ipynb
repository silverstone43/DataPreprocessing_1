```python
data = {
    '나이' : [22,35,14,55,42]
}

df = pd.DataFrame(data)

def categorize(data):
    if 0<data<-10:
        return "10대"
    elif 10<data<=20:
        return "20대"
    elif 20<data<=30:
        return "30대"
    else:
        return "30대이상"

df["검사나이"]= df["나이"].apply(categorize)
df
```

```python
df = pd.DataFrame({
    "문항1":['1','2'," ",'2'],
    "문항2":[" ",'2','3'," "]
})
df
df[" 문항 "]=df["문항1"].fillna(" ")+","+df["문항2"].fillna(" ")
df[" 문항 "]=df["문항1"].str.rstrip(", ")
df[" 문항 "]=df["문항1"].str.lstrip(",")
df=df.drop(columns=["문항1", "문항2"])
print(df)
```

```python
df = pd.DataFrame({
    "문항1":['1','2'," ",'2'],
    "문항2":[" ",'2','3'," "],
    "문항3":[" ",'1','2',""]
})

cols=["문항1","문항2","문항3"]
df[cols]=df[cols].replace(r'^\s*$','무응답',regex=True)

df["문항"]=df["문항1"]+","+df["문항2"]+","+df["문항3"]

print(df)
```
