```python
import pandas as pd

df = pd.read_csv("program_1.csv", encoding = "utf-8-sig")
print(df.head())
```

```python
num_df = df.select_dtypes(include='number')

num_df = num_df.fillna(num_df.mean())

num_df
```

```python
obj_df = df.select_dtypes(include='object')
```

```python
result = pd.concat([obj_df,num_df], axis=1)

result
```

```python
result["평균"] = (result['프로그램1'] + result['프로그램2'] + result['프로그램3'] + result['프로그램4'])/4
result
```

```python
def categorize(result):
    if 0<result<=55:
        return "하"
    elif 55<result<=70:
        return "중"
    elif 71<result:
        return "상"

result["평가분류"]= result["평균"].apply(categorize)
result
```

```python
filter_result_contains = result[result['평가분류'].str.contains('상', na=False)]
filter_result_contains
```
